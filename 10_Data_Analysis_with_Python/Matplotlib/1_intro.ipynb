### Data  Visualization Wiht Matplotlib
Metplotlib is a powerfull plotting library for python  that enables the creation  of static, animatedd, and  interactive visualization. It is widely used for data  visualization in data science and analytics.

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [1,4,9,16,25]

```

```python
##create  aline plot
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y Axis')
plt.title("Basic line Plot")

plt.show() 
```

```python
x = [1,2,3,4,5,6]
y = [1,3,5,7,9,11]

## create customized line plot
plt.plot(x,y,color='red', linestyle='--')
```

```python
plt.plot(x,y,color='red', linestyle='--',  marker='o')
```

```python
plt.plot(x,y,color='red', linestyle='--', marker='o', linewidth=2)
```

```python
plt.plot(x,y,color='red', linestyle='--', marker='o', linewidth=2, markersize=9)
plt.grid = True
```

```python
##Mutiple  Plots
##  sample data
x = [1,2,3,4,5]
y1 = [1,3,5,7,9]
y2 = [2,4,6,8,10]


plt.figure(figsize=(9,5))
plt.subplot(1,2,1)
plt.plot(x,y1,color='green')

plt.title("Plot-1")
plt.subplot(1,2,1)
plt.plot(x,y1,color='green')

plt.title("Plot-1")
```
