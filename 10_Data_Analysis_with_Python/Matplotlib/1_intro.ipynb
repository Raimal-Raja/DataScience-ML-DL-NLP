### Data  Visualization Wiht Matplotlib
Metplotlib is a powerfull plotting library for python  that enables the creation  of static, animatedd, and  interactive visualization. It is widely used for data  visualization in data science and analytics.

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [1,4,9,16,25]

```

```python
##create  aline plot
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y Axis')
plt.title("Basic line Plot")

plt.show() 
```
