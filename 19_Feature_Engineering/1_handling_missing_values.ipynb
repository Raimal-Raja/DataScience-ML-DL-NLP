### Missing Values

missing values occurs in dataset when some of the informations is not stored for a variable. There are three mechanism

## Missing Completely a Random, MCAR:
missing completely at random (MCAR) is a type of missning data mechanism in which the probability of a valu being missing is unrelated to both  the observed data and the missing data. in other words, if the  data is MCAR the missing values are randomly distributed througout the  dataset, and there is no systematic reason for why they are missing.

for example in a survey about the prevalence of  a certain disease, the missing data might be MCAR if the survey participant with missing values for certain questions were selected randomly and their missing responses are not related to the theit disease status or any other variables  measured in the  survey.


## Missing at Random MAR:
Missing at random is a type of missing data mechanism in which the probability of a value being  missing depends only on the  observed data,  but not on the missing data itself. in other words, if the data is MAR, the  missing values  are systematically related to the observed data but not the the missing data. Here are a few examples of missing at random.

income data suppose you are collecting income data from a group of people but  some participants choose not  to report their income. if the decision to report or not report income  is related to the participant's age or gender, but not to their income leve, then the  data is missing at random

Medical data: suppose you are collecting medical data on patients including their blood pressure, but some patients do not report their blood pressure, if the patients who do not report their blood pressure are more likey to be younger or healthier lifestyle but  the missingness is not related to their actual blood pressure values, then the data is missing at random.


## Missing data not at random (MNAR):
it is a type missing data mechanism where the probability of missing values depends on the values of the missing  data itself. in other words, if data is MNAR, the  missingness is not random and is dependent on unobserved or unmeasured factors that are assiociated with missing values.

for example: suppose you are collecting data on the income and job satification of employees in a company. if employees who are less satified with jobs are more likely to refuse to report their income. then the data is not missing at random in this case, the missingness is dependent on job satification, which is not directly observed or measured.
